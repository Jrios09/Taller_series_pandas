```python
#Aplicaciones aritmeticas
import pandas as pd
# Crear dos Series numéricas
serie1 = pd.Series([10, 20, 30, 40])
serie2 = pd.Series([5, 10, 15, 20])

# 2. Restar una Serie de otra
print("\nResta:")
print(serie1 - serie2)

# 3. Multiplicar los valores de una Serie por 2
print("\nMultiplicación por 2:")
print(serie1 * 2)

# 4. Dividir los valores de una Serie entre 2
print("\nDivisión entre 2:")
print(serie1 / 2)

# 5. Operación combinada
print("\nOperación combinada (serie1 + serie2) / 2:")
print((serie1 + serie2) / 2)
```

```python
#aplicacion de funciones
import pandas as pd
import numpy as np

# 1. Lambda para elevar al cuadrado
serie_num = pd.Series([1, 2, 3, 4, 5])
cuadrado = serie_num.apply(lambda x: x**2)
print("Cuadrado:")
print(cuadrado)

# 2. map() para convertir nombres a mayúsculas
nombres = pd.Series(["juan", "ana", "luis"])
mayusculas = nombres.map(str.upper)
print("\nNombres en mayúsculas:")
print(mayusculas)

# 3. apply(np.sqrt)
raiz = serie_num.apply(np.sqrt)
print("\nRaíz cuadrada:")
print(raiz)

# 4. Función personalizada con apply()
def clasificar(x):
    if x >= 3:
        return "Alto"
    else:
        return "Bajo"

clasificacion = serie_num.apply(clasificar)
print("\nClasificación:")
print(clasificacion)

# 5. map() con diccionario
numeros = pd.Series([1, 2, 3, 1, 2])
dic = {1: "Uno", 2: "Dos", 3: "Tres"}
reemplazo = numeros.map(dic)

print("\nReemplazo con diccionario:")
print(reemplazo)
```

```python
#Ordenacion y clasificacion
import pandas as pd

# Serie numérica
serie_num = pd.Series([45, 12, 78, 34, 23, 90])

# 1. Ordenar de forma ascendente
orden_asc = serie_num.sort_values()
print("Orden ascendente:")
print(orden_asc)

# Serie de texto
serie_texto = pd.Series(["banana", "manzana", "pera", "uva"])

# 2. Ordenar texto alfabéticamente
orden_texto = serie_texto.sort_values()
print("\nTexto ordenado alfabéticamente:")
print(orden_texto)

# Serie con índice desordenado
serie_indice = pd.Series([100, 200, 300], index=['c', 'a', 'b'])

# 3. Ordenar por índice
orden_indice = serie_indice.sort_index()
print("\nOrden por índice:")
print(orden_indice)

# 4. Tres valores más altos
mayores = serie_num.nlargest(3)
print("\nTres valores más altos:")
print(mayores)

# 5. Tres valores más bajos
menores = serie_num.nsmallest(3)
print("\nTres valores más bajos:")
print(menores)
```

```python
#Filtracion y seleccion condicional
import pandas as pd

# Serie numérica
serie = pd.Series([10, 25, 40, 55, 60, 75, 90])

# 1. Filtrar valores mayores que 50
mayores_50 = serie[serie > 50]
print("Valores mayores que 50:")
print(mayores_50)

# 2. Mostrar los valores pares
pares = serie[serie % 2 == 0]
print("\nValores pares:")
print(pares)

# Serie de texto
serie_texto = pd.Series(["python", "java", "python avanzado", "c++", "aprendiendo python"])

# 3. Filtrar valores que contengan la palabra "python"
contiene_python = serie_texto[serie_texto.str.contains("python")]
print("\nTextos que contienen 'python':")
print(contiene_python)

# 4. Usar condición doble (>30 y <70)
entre_30_70 = serie[(serie > 30) & (serie < 70)]
print("\nValores entre 30 y 70:")
print(entre_30_70)

# 5. Obtener los índices de los valores válidos (>50)
indices_validos = serie[serie > 50].index
print("\nÍndices de valores mayores que 50:")
print(indices_validos)
```

```python
#Manejo de valores nulos

import pandas as pd
import numpy as np

# Crear una Serie con valores nulos
serie = pd.Series([10, 20, np.nan, 40, np.nan, 60])

print("Serie original:")
print(serie)

# 1. Reemplazar los NaN con 0
reemplazo_cero = serie.fillna(0)
print("\nReemplazar NaN con 0:")
print(reemplazo_cero)

# 2. Eliminar los valores nulos
sin_nulos = serie.dropna()
print("\nSerie sin valores nulos:")
print(sin_nulos)

# 3. Contar valores nulos
cantidad_nulos = serie.isna().sum()
print("\nCantidad de valores nulos:")
print(cantidad_nulos)

# 4. Llenar los nulos con el promedio de la Serie
rellenar_promedio = serie.fillna(serie.mean())
print("\nNaN reemplazados con el promedio:")
print(rellenar_promedio)

# 5. Reemplazar los nulos con un valor constante (ej: 999)
reemplazo_constante = serie.fillna(999)
print("\nNaN reemplazados con valor constante:")
print(reemplazo_constante)
```

```python
#Operacion de alineacion y combinacion
import pandas as pd

# Crear dos Series con índices distintos
serie1 = pd.Series([10, 20, 30], index=['a','b','c'])
serie2 = pd.Series([5, 15, 25], index=['b','c','d'])

print("Serie 1:")
print(serie1)

print("\nSerie 2:")
print(serie2)

# 1. Sumar dos Series con índices distintos
suma = serie1 + serie2
print("\nSuma de Series:")
print(suma)

# 2. Combinar usando add() con fill_value=0
suma_fill = serie1.add(serie2, fill_value=0)
print("\nSuma con fill_value=0:")
print(suma_fill)

# 3. Usar combine_first() para fusionar Series
fusion = serie1.combine_first(serie2)
print("\nFusion con combine_first():")
print(fusion)

# 4. Ver cómo Pandas alinea los datos automáticamente
print("\nAlineación automática de Pandas:")
print(serie1 + serie2)

# 5. Operación aritmética compleja
operacion_compleja = (serie1.add(serie2, fill_value=0)) * 2
print("\nOperacion aritmetica compleja:")
print(operacion_compleja)
```

```python
#Reindexacion
import pandas as pd
import numpy as np

# Crear una Serie base
serie = pd.Series([10, 20, 30, 40], index=['a','b','c','d'])

print("Serie original:")
print(serie)

# 1. Reindexar agregando un nuevo índice
reindex1 = serie.reindex(['a','b','c','d','e'])
print("\n1. Reindex con nuevo índice 'e':")
print(reindex1)

# 2. Reindexar en orden inverso
reindex2 = serie.reindex(['d','c','b','a'])
print("\n2. Serie en orden inverso:")
print(reindex2)

# 3. Reindex con fill_value
reindex3 = serie.reindex(['a','b','c','d','e'], fill_value=0)
print("\n3. Reindex con fill_value=0:")
print(reindex3)

# 4. Rellenar valores faltantes con el promedio
reindex4 = serie.reindex(['a','b','c','d','e'])
promedio = serie.mean()
reindex4 = reindex4.fillna(promedio)

print("\n4. Rellenar faltantes con el promedio:")
print(reindex4)

# 5. Reindex con índices nuevos y existentes
reindex5 = serie.reindex(['b','c','e','f'])
print("\n5. Reindex con índices nuevos y existentes:")
print(reindex5)
```
